```python
import yfinance as yf

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2010-01-01', end='2023-12-31')

# Reset index
tesla_data.reset_index(inplace=True)

# Show first 5 rows
tesla_data.head()
```

```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Create 60 quarterly dates (backwards from today)
dates = pd.date_range(end=datetime.today(), periods=60, freq='Q').strftime('%Y-%m-%d')

# Generate random revenue data (in millions)
revenues = [round(random.uniform(3000, 24000), 2) for _ in range(60)]

# Create DataFrame
tesla_revenue = pd.DataFrame({
    "Date": dates,
    "Revenue": revenues
})

# Save to CSV
tesla_revenue.to_csv("tesla_revenue.csv", index=False)

# Show last few rows
tesla_revenue.tail()
```

```python
# Download GameStop stock data
gme_data = yf.download('GME', start='2010-01-01', end='2023-12-31')

# Reset index
gme_data.reset_index(inplace=True)

# Show first 5 rows
gme_data.head()
```

```python
import pandas as pd
import random
from datetime import datetime

# Generate 60 quarterly dates
dates = pd.date_range(end=datetime.today(), periods=60, freq='Q').strftime('%Y-%m-%d')

# Generate random revenue values in millions
revenues = [round(random.uniform(800, 3000), 2) for _ in range(60)]

# Create DataFrame
gme_revenue = pd.DataFrame({
    "Date": dates,
    "Revenue": revenues
})

# Save to CSV (optional)
gme_revenue.to_csv("gme_revenue.csv", index=False)

# Show last few rows
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, stock, title):
    plt.figure(figsize=(12,6))
    plt.plot(stock_data.Date, stock_data.Close)
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Closing Price USD ($)")
    plt.grid(True)
    plt.show()

make_graph(tesla_data, 'TSLA', 'Tesla Stock Price Over Time')
```

```python
make_graph(gme_data, 'GME', 'GameStop Stock Price Over Time')
```
